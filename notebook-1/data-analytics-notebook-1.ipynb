### StackPlot - Brisbane CityCycle — Membership and Trip Numbers

---

Question: How have membership and trip numbers changed in the last six months of 2018?

---
*This data file or API can be downloaded from:
https://www.data.brisbane.qld.gov.au/data/dataset/773a765d-aaac-4f8a-8ac3-1617dfb09fbd/resource/40b7d7a2-4a06-427b-b752-4999511c637c/download/citycycle-memberships-and-trip-numbers-31.12.2018.csv - Creative Commons Attribution 4.0 Australia*

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
cyclefile = 'brisbane-citycycle.csv' #identify file
cycle_data = pandas.read_csv(cyclefile)
```

```python
cycle_data #original file
```

```python
cycle_data_2018 = cycle_data.iloc[-6:] #locate the last 6 rows
cycle_data_2018 #locate the last 6 months
```

```python
cycle_data_key_2018 = cycle_data_2018.filter(['Month', 'Total_Trips', 'Members_Short-Term_Total', 'Members_Long-Term_Total_Sign_ups']) #filter by key columns
cycle_data_key_2018 #filer key variables
```

```python
x = cycle_data_key_2018['Month'] #define x axis
y = np.vstack([cycle_data_key_2018['Total_Trips'], cycle_data_key_2018['Members_Short-Term_Total'],cycle_data_key_2018['Members_Long-Term_Total_Sign_ups']]) #define y axis
```

```python
labels = ['Total Trips', 'Short Term Memberships', 'Long-Term Sign-Ups'] #define labels
axes = plt.gca() #set y axis limit
axes.set_ylim([58000,83000]) #set y axis limit
plt.stackplot(x, y, labels = labels, edgecolor = 'black') #create stackplot
plt.legend(loc = 'lower left', bbox_to_anchor = (1, 0.1)) #legend location
plt.show()

#colors = ['light purple', 'green', 'goldenrod'] - extra colors to try out
```

#### Answer: How have membership and trip numbers changed in the last six months of 2018?

1. Total trips were declining between July to October, but had a brief spike in November
2. Long-term sign-up rates have remained relatively stable
3. Short-term memberships experienced a short lull but have been improving since October

### Line Graph - Sea Surface Temperature

---

Question: How has sea surface temperature anomaly changed over the last 10 years?

---
*This dataset was originally found on Queensland Government
https://data.qld.gov.au/dataset/9159a149-dac0-4f7c-8548-5714f0e9c274 - Creative Commons Attribution 3.0*

```python
import matplotlib.pyplot as plt
```

```python
seafile = 'sea-surface-temperature.csv' #identify file
seacolname = 'Year' #column name 
sea_data = pandas.read_csv(seafile, index_col = seacolname) #read file and identify index column 
```

```python
sea_data #original file
```

```python
sea_data_surf = sea_data.loc['Sea surface temperature anomaly', '2007':'2017'] #filter years
sea_data_rol = sea_data.loc['Ten year rolling average', '2007':'2017'] #filter years
plt.plot(sea_data_surf,'g-', linestyle = "--",label ='Sea surface temperature anomaly') #colour, linestyle, and label
plt.plot(sea_data_rol,'m-', linestyle = "-.", label ='Ten year rolling average') #colour, linestyle, and label
plt.ylabel('Data') #y axis label
plt.xlabel('Year') #x axis label
plt.suptitle('Sea temperature changes') #title 
plt.legend(loc = 'lower right', bbox_to_anchor = (1.8, 0.5)) #location of legend 
plt.grid(True) #grid 
plt.savefig('seaplot.jpg') #save as .jpg
```

#### Answer: How has sea surface temperature anomaly changed over the last 10 years?

1. There has been a spike in the sea surface temperature over the last few years, followed by a short decline
2. The ten year rolling average has increased over time
